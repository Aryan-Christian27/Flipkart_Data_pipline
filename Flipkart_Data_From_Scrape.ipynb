```python
# Function to extract Availability Status
def get_availability(soup):
    try:
        # Locate the main availability div
        available = soup.find("div", class_="M4DNwV")

        # Check if the first inner div exists
        if available:
            available = available.find("div", class_="n5vj9c")

        # Extract text from the deepest availability div
        if available:
            available = available.find("div", class_="yiggsN O5Fpg8").text.strip()

    except AttributeError:
        available = "Not Available"	

    return available
```
